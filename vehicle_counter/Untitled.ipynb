```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
model = YOLO("yolov8n.pt")
```

```python
cap = cv2.VideoCapture("highway.mp4")
while cap.isOpened():
    ret, frame = cap.read()
    if ret:
        results = model.track(frame, show=True)
        if cv2.waitKey(33) == ord("q"):
            break
    else:
        break
cap.release()
cv2.destroyAllWindows()
```
